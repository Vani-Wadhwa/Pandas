```python
import pandas as pd
pd.Series([[1,2,3],[4,5,6]])
```

```python
data={'subject':[1,2,3,4,5],'first_name':['A','B','C','D','E'],'last_name':['F','G','H','I','J']}
```

```python
data
```

```python
df1=pd.DataFrame(data,columns=['subject','first_name','last_name'])
df1
```

```python
data1={'subject':[4,5,6,7,8],'first_name':['K','L','M','N','O'],'last_name':['P','Q','R','S','T']}
```

```python
df2=pd.DataFrame(data1,columns=['subject','first_name','last_name'])
df2
```

CONCATENATION - It basically gives the dataframes together. Concat 2 dataframes along the rows, by default axis=0 that is rows

```python
dfnew=pd.concat([df1,df2])
dfnew
```

CONCATENATION along columns

```python
pd.concat([df1,df2],axis=1)
```

```python
data3={'subject':[1,2,3,4,5,6,7,8],"test_id":[51,52,53,54,55,56,57,58]}
df3=pd.DataFrame(data3,columns=['subject','test_id'])
df3
```

MERGE - Merge 2 dataframes to form a single dataframe.

```python
df_new=pd.merge(df3,dfnew)
```

```python
df_new
```

Merge with both left and right dataframes using the subject key

```python
df_new=pd.merge(df3,dfnew,on="subject")
```

```python
df_new
```

FULL OUTER JOIN PRODUCES THE SETS OF ALL RECORD IN TABLE 1 AND TABLE 2, WITH MATCHING RECORDS FROM BOTH SIDES WHERE AVAILABLE

```python
pd.merge(df1,df2,on='subject',how='outer')
```

INNER JOIN PRODUCES ONLY THE SET OF RECORDS THAT MATCH IN BOTH TABLE 1 AND TABLE 

```python
pd.merge(df1,df2,on='subject',how='inner')
```

APPEND FUNCTION - DATA WILL BE ADDED ALONG THE ROWS

```python
df1.append(df2)
```

```python
df_c=pd.DataFrame({"HPI":[90,80,70,60],"unemployment":[5,4,6,4]},index=[2001,2002,2003,2004])
df_d=pd.DataFrame({"Int_rate":[2,3,2,1],"Ind_GDP":[40,60,70,70]},index=[2001,2003,2004,2005])
```

JOIN - combine 2 differently indexed dataframes into a single dataframe. This is quite similar to merge operation, expect the joing operation will be on the "index" instead of "columns"

```python
df_c.join(df_d)
```

```python
df_d.join(df_c)
```
